```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_excel('C:\\Users\\jains\\Downloads\\Data_Train.xlsx')
df.head()
```

```python
df.info()
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis = 0,inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
stop_mapping = {'non-stop' : 0,'1 stop': 1,'2 stops':2,'3 stops':3,'4 stops':4}
df['Total_Stops'] = df['Total_Stops'].map(stop_mapping).astype('int')
df.head()
```

```python
df['Additional_Info'].unique()
```

```python
df['Journey_month'] = df['Date_of_Journey'].dt.month.astype('int')
df['Journey_day'] = df['Date_of_Journey'].dt.day.astype('int')
df.drop(['Date_of_Journey'],axis = 1,inplace = True)
df.head()
```

```python
df['Dep_hr'] = pd.to_datetime(df['Dep_Time'],format='%H:%M').dt.hour.astype('int')
df['Dep_min'] = pd.to_datetime(df['Dep_Time'],format='%H:%M').dt.minute.astype('int')
df['Arr_hr'] = df['Arrival_Time'].apply(lambda x:pd.to_datetime(x,format='%H:%M').hour if len(x) == 5 else pd.to_datetime(x,format='%H:%M %d %b').hour).astype('int')
df['Arr_min'] = df['Arrival_Time'].apply(lambda x:pd.to_datetime(x,format='%H:%M').minute if len(x) == 5 else pd.to_datetime(x,format='%H:%M %d %b').minute).astype('int')
df.drop(columns=['Arrival_Time','Dep_Time'],axis =1,inplace = True)
df.head()
```

```python
def convertInMinutes(duration):
    hour,min = 0,0
    if 'h' in duration:
        hour = int(duration.split('h')[0].strip())
        duration = duration.split('h')[1].strip()
    if 'm' in duration:
        min = int(duration.split('m')[0].strip())
    return 60*hour + min
df['Duration'] = df['Duration'].apply(convertInMinutes).astype('int')
```

```python
df.head()
```

```python
df['Airline'].unique()
```

```python
df['Airline'] = df['Airline'].astype('category').cat.codes
df['Source'] = df['Source'].astype('category').cat.codes
df['Destination'] = df['Destination'].astype('category').cat.codes
df['Additional_Info'] = df['Additional_Info'].astype('category').cat.codes
```

```python
x_train = df.drop(['Route','Price'],axis =1)
y_train = df['Price']
```

```python
X_test = x_train.tail(2672)
y_test = y_train.tail(2672)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train,y_train)
predictions = model.predict(X_test)
print(f"Mean Absolute Error: {mean_absolute_error(y_test, predictions)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, predictions)}")
print(f"R^2 Score: {r2_score(y_test, predictions)}")
```
